```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
```

```python
df = pd.read_html('https://www.numbeo.com/cost-of-living/prices_by_city.jsp?displayCurrency=USD&itemId=105', header=0)[3]
df.columns = ['Rank', 'City Nation', 'Salary']
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.iloc[:, 1:3]
```

```python
is_US = df['City Nation'].str.contains('United States')
is_China = df['City Nation'].str.contains('Qingdao, Shandong, China')
is_Canada = df['City Nation'].str.contains('Nanaimo, BC, Canada')
is_Georgia = df['City Nation'].str.contains('Georgia') & ~df['City Nation'].str.contains('Tbilisi')

df_US = df[is_US]
df_China = df[is_China]
df_Canada = df[is_Canada]
df_Georgia = df[is_Georgia]
df_not_US = df[~is_US & ~is_China & ~is_Canada & ~is_Georgia]

for d in [df_US, df_China, df_Canada, df_Georgia]:
#     print(d.head())
    d[['City', 'State', 'Nation']] = d['City Nation'].str.split(',', expand=True)

df_not_US[['City', 'Nation']] = df_not_US['City Nation'].str.split(',', expand=True)
df_not_US['State'] = 'NA'
```

```python
df_clean = pd.concat([df_US, df_China, df_Canada, df_Georgia, df_not_US])
```

```python
df_clean.shape
```

```python
df_clean[df_clean['Nation'] == 'NA']
```

## Data Exploration

```python
df_clean['Global_Ranking'] = df_clean['Salary'].rank(ascending=False)
df_clean['National_Ranking'] = df_clean.groupby(['Nation'])['Salary'].rank(ascending=False)
df_clean['Statewise_Ranking'] = df_clean.groupby(['Nation', 'State'])['Salary'].rank(ascending=False)
```

```python
#calculate AVG(value) OVER (PARTITION BY group)
df_national_ranking = df_clean[(df_clean['National_Ranking'] == 1) | 
                               (df_clean['City'] == 'Philadelphia')
                              ].sort_values(['Salary'], ascending = False).reset_index().head(50)
df_national_ranking.head(20)
```

```python
plt.rcParams['figure.figsize'] = [8, 16]
fig, ax = plt.subplots()
y_pos = np.arange(len(df_national_ranking))

ax.barh(y_pos, df_national_ranking['Salary'])
ax.set_yticks(y_pos)
ax.set_yticklabels(df_national_ranking['City Nation'])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Salary')

plt.show()
```

```python
df_national = df_clean.groupby(['Nation']).mean().reset_index()
df_national['Rank_by_Nation'] = df_national['Salary'].rank(ascending=False)
df_national.sort_values(['Rank_by_Nation']).head(10)
```

```python
df_clean = df_clean.merge(df_national[['Nation', 'Rank_by_Nation']], 
                          left_on='Nation', right_on='Nation').sort_values(['Rank_by_Nation'])
df_clean.head()
```

```python
df_viz = df_clean[df_clean['Rank_by_Nation'] <= 40]
sns.boxplot( y=df_viz["Nation"], x=df_viz["Salary"] )
plt.axvline(df_viz[df_viz['City'] == 'Philadelphia']['Salary'].values)
plt.show()
```
